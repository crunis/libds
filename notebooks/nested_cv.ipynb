```python
from sklearn.datasets import load_iris
import pandas as pd

iris_data = load_iris()

X = iris_data.data
y = iris_data.target
X2 = X[y<2]
y2 = y[y<2]

df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
df['target'] = iris_data.target

# print(df.head())
```

```python
rx = XGBNestedCV(balance_classes = True, stratified=False, trials=35, timeout=60, max_features=4, verbose=0)
```

```python
data = rx.go(X2, y2, iris_data.feature_names)
```
